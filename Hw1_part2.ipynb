```python
import numpy as np
import scipy as sp 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
pd_data = pd.read_csv("./who_suicide_statistics.csv")
pd_data.head(5)
```

```python
suicide_ratio =  100*pd_data.suicides_no/pd_data.population
suicide_ratio
```

```python
pd_data.insert(1, 'suicide_ratio', suicide_ratio)
pd_data
```

```python
#sns.barplot(x = "year", y = suicide_ratio, data = pd_data).set_title("Years vs Suicide rates")
ax = pd_data[['year', 'suicide_ratio']].groupby(['year'])
ax.mean().tail(15)
```

```python
ax.mean().tail(15).plot.bar()
```

* 可以觀察到近幾年死亡率的變化

```python
abc = pd_data[['sex', 'age', 'suicide_ratio']].groupby(['sex', 'age'])
abc.mean()
```

```python
abc.mean().plot.bar()
```

* 根據不同男女與年紀的分佈觀察，男性自殺率高於女性

```python
country = pd_data[['country', 'suicide_ratio']].groupby(['country'])
#country.sort_values(by=['suicide_ratio'])
country.mean().sort_values(by=['suicide_ratio'], ascending=False).head(10)
```

```python
country.mean().sort_values(by=['suicide_ratio'], ascending=False).head(10).plot.bar()
```

* 依照自殺率由高到低的國家來排列，也可以顯示中間的國家只要將 `head(10)` 改成 `[a:b]` 可顯示排名 a~b 的國家
